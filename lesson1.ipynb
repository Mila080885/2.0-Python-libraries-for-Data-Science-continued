### 1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Максимизировать лучше Precision, когда высока "цена" ложно-положительных результатов. Например, когда есть этические вопросы или когда дорогостоящиа или трудозатратная проверка прогнозируемых результатов.

Recall улучшается - когда высок риск не выявить нужные признаки, например, мошейнические операции, диагностика серьезных заболеваний

### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Часто в реальной практике стоит задача найти оптимальный (для заказчика) баланс между этими двумя метриками. 

F-мера представляет собой гармоническое среднее между точностью и полнотой. Она стремится к нулю, если точность или полнота стремится к нулю.

В статистике среднее гармоническое применяется в случае, когда наблюдения, для которых требуется получить среднее арифметическое, заданы обратными значениями.

Данная формула придает одинаковый вес точности и полноте, поэтому F-мера будет падать одинаково при уменьшении и точности и полноты. Возможно рассчитать F-меру придав различный вес точности и полноте

 β принимает значения в диапазоне 0<β<1 если вы хотите отдать приоритет точности, а при β>1 приоритет отдается полноте.
