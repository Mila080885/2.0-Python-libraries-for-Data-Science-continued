### Урок 2. Анализ данных и проверка статистических гипотез

Ответ Цымбал Людмилы

#### 1. В чём различие между зависимыми и независимыми выборками?

Начну с понятия Выборка. 
**Выборка** — это ограниченная по численности группа объектов, специально отбираемая из генеральной совокупности для изучения ее свойств. **Зависимые выборки** 1. Выборки, характеризующиеся тем, что каждому объекту одной выборки поставлен в соответствие по определенному критерию объект из другой выборки. 2. Допущение зависимости чаще всего означает, что признак измерен на одной и той же выборке дважды, например, до воздействия и после него. Более слабые виды зависимости выборок: выборка 1 – мужья, выборка 2 – их жены. **Независимые выборки** 1. Выборки, которые характеризуются тем, что вероятность отбора любого объекта одной выборки не зависит от отбора любого объекта из другой выборки. 2. Допущение независимости предполагает, что представители двух выборок не составляют пары коррелирующих значений признака.

#### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

**Параметрические – критерии**, основанные на оценке параметров распределения, к которым относятся среднее арифметическое, среднеквадратическое отклонение, дисперсия. Они  применимы только в том случае, если численные данные подчиняются нормальному распределению. Если распределение отличается от нормального, то следует пользоваться так называемыми непараметрическими критериями.

**Непараметрические критерии** не основаны на оценке параметров распределения и вообще не требуют, чтобы данные подчинялись какому-то определенному типу распределения.  Непараметрические критерии дают более грубые оценки, чем параметрические, но являются более универсальными. А параметрические методы более точны, но лишь в случае, если правильно определено распределение совокупности.

#### Виды критериев (наиболее часто используемые):
    
**Параметрические методы**

Если  выборки из нормальных совокупностей (нормальное распределение) с одинаковыми генеральными дисперсиями.
**r-Пирсона** - предназначен для оценки связи (линейной) между двумя переменными, измеренными в метрической шкале, распределение которых соответствует нормальному (иногда дает достаточно точные результаты и в случаях аномальных распределений, а также в случаях, когда одна из переменных является дискретной (состоящая из ряда качественно отличных друг от друга и раздельных единиц или категорий, но для распределений, не являющихся нормальными, предпочтительнее пользоваться ранговыми коэффициентами корреляции Спирмена или Кендалла.)).
**T-критерий Стъюдента**: направлен на оценку различий средних величин двух выборок распределённых по нормальному закону. Используется для связных и несвязных выборок, которые могут быть неравны по величине. Позволяет оценить сдвиг значений признака и выявить различия в его распределении.
**F-критерий Фишера** позволяет сравнивать величины выборочных дисперсий двух рядов наблюдений.

**Непараметрические методы**

Если обе переменные, между которыми изучается связь, представлены в порядковой шкале (ранговой), или одна из них - в порядковой, а другая - в метрической, или для метрических, когда связь между переменными является нелинейной
**Спирмена** или **τ-Кенделла** - предназначены для оценки связи между двумя переменными, измеренными

Если две независимые выборки кач. и кол. шкал:
Критерий **U-Манна-Уитни** (выборка от 3 до 60). Применяется для оценки различий по показателям какого-либо признака. Количество показателей в выборках может быть неодинаковым. Чем больше различий, тем меньше эмпирическое значение U, тем более вероятно, что различия достоверны.

Если две зависимые выборки кач. и кол. шкал:
Критерий **знаков G** (анализ общего направления сдвига значений – от 5 до 300). Позволяет установить, в какую сторону в выборке в целом изменяются значения признака при переходе от первого измерения ко второму; 
Критерий **Т-Вилкоксона** (парный) (анализ интенсивности сдвига значений – от 5 до 50). Применяется для сопоставления показателей, измеренных в двух разных условиях на одной и той же выборке испытуемых. Позволяет установить не только направленность изменений, но и их выраженность. С его помощью мы определяем, является ли сдвиг показателей в каком-то направлении более интенсивным, чем в другом.

Если более двух независимых выборок кач. и кол. шкал:
Критерий **H–Крускала-Уоллиса** (не менее 3, 4 наблюдений), без конкретизации направления различий. Для утверждения о том, что уровень выраженности признака в какой-то из сравниваемых выборок выше или ниже

Если более двух зависимых выборок кач. и кол. шкал:
Критерий **Фридмана**- применяется для сопоставления показателей, измеренных в трех или более условиях на одной и той же выборке испытуемых. Критерий позволяет установить, что величины показателей от условия к условию изменяются, но при этом не указывает на направление изменений.

Если две или более номинативные (категориальные) переменные:
**Хи-квадрат Пирсона** - используется для проверки нулевой гипотезы о подчинении наблюдаемой случайной величины определенному теоретическому закону распределения. Используется, когда необходимо установить соответствие двух сравниваемых рядов распределения у - эмпирического и теоретического, или двух эмпирических. При этом сравниваются частоты названных рядов распределения, выявляются расхождения между ними и определяется вероятность этих расхождений. Представляет собой некоторую величину, которая оценивается с определенной вероятностью. Он может принимать различные всегда положительные значения (малые и большие). Данный критерий не рекомендуется использовать для оценки малых выборок (желательно >50). Значение параметра возрастает с увеличением разницы между частотами.
